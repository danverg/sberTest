```python
#Загружаем выгрузку из прошлого пункта
data = pd.read_csv('GeoData.csv', delimiter=',')
geo = geo.merge(data[["ref", 'counter', 'perc']])
geo
```

```python
import folium
import json
import geojson as gj
m = folium.Map(location=[55.75389793422123, 37.6207653732435], zoom_start=8.5)

#Создаем шаблон popup
def popup_html(feature):
    html = '<h5>{}</h5>'.format(feature['properties']['name'])
    html += '<br><b>Процент заказов</b>: {:.3f}%'.format(feature['properties']['perc'])
    html += '<br><b>Количество заказов</b>: {}'.format(feature['properties']['counter'])
    return html

#Задаем стиль накладываемого слоя
def DefStyle(feature):
    return {"fillColor":'white', "fillOpacity":0,"opacity":0}

#Описываем изменение при наведении
def HoverStyle(feature):
    return {"fillColor":'black', "fillOpacity":0.1,"opacity":0}

#Создаем хороплет
rel = folium.Choropleth(
       geo_data = geo.to_json(), 
       name = 'Количество заказов в МСК',
       data = data,
       columns=['ref', 'perc'], 
       key_on="feature.properties.ref",
       #bins = 3,
       fill_color='BuGn',
       nan_fill_color='white',
       nan_fill_opacity=0.5,
       fill_opacity=0.7,
       line_opacity=0.2,
       legend_name= 'Процент заказов в МСК',
       highlight = True,
       show = False,
)
rel.add_to(m)

#Добовляем popups
for mo in json.loads(geo.to_json())['features']:
    reg = folium.GeoJson(data=mo, style_function = DefStyle, highlight_function= HoverStyle,overlay=False, control=False, show=False, smooth_factor=None, tooltip=None, embed=False, popup=None, zoom_on_click=False, marker=None)
    folium.Popup(popup_html(mo)).add_to(reg)
    reg.add_to(rel)
m
```

```python
m.save('map.html')
```
